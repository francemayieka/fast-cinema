```python
import psycopg2
```

```python
import os
from dotenv import load_dotenv

# Load the .env file (make sure it's in the same directory as this notebook)
load_dotenv()
DATABASE_URL = os.getenv('DATABASE_URL')

# Fail fast if the database connection string wasn't loaded securely
if not DATABASE_URL:
    raise ValueError("DATABASE_URL not found in .env file!")
```

```python
conn = psycopg2.connect(DATABASE_URL)
```

```python
cur = conn.cursor()
cur.execute("SELECT version();")
```

```python
cur.fetchone()
```

```python
cur.execute("SELECT * FROM movies LIMIT 5;")
cur.fetchall()
```

```python
# ensure any previous failed transaction is cleared
conn.rollback()
```

```python
# select id, title from movies where release_year > 2000;
cur.execute("SELECT movie_id, title FROM movies WHERE release_year > 2000 limit 5;")
cur.fetchall()
```

```python
# count
cur.execute("SELECT COUNT(*) FROM movies;")
cur.fetchone()
```

```python
# average rating
cur.execute("SELECT AVG(rating) FROM movies;")
cur.fetchone()
```

```python
import pandas as pd
df = pd.read_sql_query("SELECT * FROM movies LIMIT 10;", conn)
df.head()
```
